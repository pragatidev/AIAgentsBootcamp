## ğŸ—ï¸ Step 5: Create Virtual Environment

Virtual environments isolate your project dependencies. Let's create one using the terminal.

### ğŸ”§ Terminal Commands:

**If UV is working (from previous step):**
```bash
uv venv .venv
```

**If UV failed, use standard Python:**
```bash
python -m venv .venv
```

### ğŸ¯ Activate Virtual Environment:

**Windows:**
```bash
.venv\Scripts\activate
```

**Mac/Linux:**
```bash
source .venv/bin/activate
```

**âœ… Success Indicator**: Your terminal prompt should show `(.venv)` at the beginning

**ğŸ’¡ VS Code Tip**: VS Code will automatically detect and use this environment

**After running the commands above, run the verification cell below:**

````python
# Run this cell AFTER creating your virtual environment
from pathlib import Path
import sys

print("ğŸ” Virtual Environment Check")
print("="*30)

venv_path = Path(".venv")
if venv_path.exists():
    print("âœ… .venv folder exists!")
    
    # Check for Python executable
    if sys.platform == "win32":
        python_exe = venv_path / "Scripts" / "python.exe"
        activate_cmd = ".venv\\Scripts\\activate"
    else:
        python_exe = venv_path / "bin" / "python"
        activate_cmd = "source .venv/bin/activate"
    
    if python_exe.exists():
        print(f"âœ… Python executable found!")
        print(f"\nğŸ’¡ To activate: {activate_cmd}")
        print("ğŸ’¡ Look for (.venv) in your terminal prompt!")
    else:
        print("âŒ Python executable missing")
else:
    print("âŒ .venv folder not found")
    print("Please run the terminal commands from Step 5")

print("\nğŸ¯ Virtual environment setup complete!")
````
